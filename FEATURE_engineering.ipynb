DETECT OUTLIERS USING IQR

```python
Q1=df.height.quantile(0.25)
Q3=df.height.quantile(0.75)
Q1,Q3
```

```python
IQR = Q3 - Q1
IQR
```

IQR





lower_limit = Q1 - 1.5*IQR

lower_limit, upper_limit

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
# This line filters the DataFrame to include only the rows where the height is less than the lower limit or greater than the upper limit.

df[(df.height<lower_limit) | (df.height>upper_limit)]
```

```python
outlier_no=df[(df.height>lower_limit) & (df.height<upper_limit)]
outlier_no
```
